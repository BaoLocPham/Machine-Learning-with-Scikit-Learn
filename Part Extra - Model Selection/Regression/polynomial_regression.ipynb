# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
dataset.isna().value_counts(), dataset.isnull().value_counts()
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Polynomial Regression model on the Training set

In short, our regerssion equation looks like this:
$$\textrm{y} = b_{0} + b_{1}\times x_{1} + b_{2}\times x_{2}^{2} + b_{3}\times x_{3}^{3} +  b_{4}\times x_{4}^{4}$$

```
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_reg = PolynomialFeatures(degree=4) # because we have 4 features
X_poly_train = poly_reg.fit_transform(X_train)

poly_linear_regressor = LinearRegression()
poly_linear_regressor.fit(X_poly_train, y_train)
```

## Predicting the Test set results

```
X_poly_test = poly_reg.transform(X_test)
y_pred = poly_linear_regressor.predict(X_poly_test)

np.set_printoptions(precision=2)
np.concatenate((y_test.reshape(-1,1), y_pred.reshape(-1,1)), axis=1)
```

## Evaluating the Model Performance

```
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
